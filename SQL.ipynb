```python
#Set SQLite and create tables
#Run this cell to initalize the database

import sqlite3
import pandas as pd


#create an inmemory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
#Create Sales Data table
cursor.execute ('''
  CREATE TABLE SalesData (
    Product TEXT,
    Region TEXT,
    Sales_Amount INTEGER,
    Units_Sold INTEGER,
    Year INTEGER
  )
''')
```

```python
#Insert data into Salesdata

Sales_data = [
    ('Laptop', 'North', 1500, 10, 2023),
    ('Laptop', 'South', 1300, 8, 2023),
    ('Smartphone', 'North', 800, 15, 2023),
    ('Smartphone', 'South', 900, 20, 2023),
    ('Tablet', 'East', 700, 10, 2024),
    ('Tablet', 'West', 750, 12, 2024),
    ('Laptop', 'East', 1600, 11, 2024),
    ('Laptop', 'West', 1550, 9, 2024),
    ('Smartphone', 'North', 850, 14, 2024),
    ('Smartphone', 'South', 950, 18, 2024),
    ('Desktop', 'East',  1250, 11, 2024),
    ('Desktop', 'West',  1350, 11, 2024)
]


cursor.executemany('INSERT INTO SalesData VALUES (?,?,?,?,?)', Sales_data)
```

```python
df = pd.read_sql_query('SELECT * FROM SalesData', conn)
df
```

```python
#Create ProductDetail Table
cursor.execute ('''
  CREATE TABLE ProductDetail (
    Product TEXT,
    Category TEXT,
    Brand TEXT
  )
''')
```

```python
#Insert data into ProductDetail Table

Product_details = [

    ('Laptop', 'Electronics', 'Dell'),
    ('Smartphone', 'Electronics', 'Samsung'),
    ('Tablet', 'Electronics', 'Apple'),
    ('Smartwatch', 'Electronics', 'Garmin'),
]

cursor.executemany('INSERT INTO ProductDetail VALUES (?,?,?)', Product_details)

conn.commit()
```

```python
df = pd.read_sql_query('SELECT * FROM Productdetail', conn)
df
```

```python
def run_query(query):
  return pd.read_sql_query(query, conn)
  return df
```

```python
#Left Join
#The tripple quotes  ('''' or """) in python  are used to define a mutli_line string

query = ('''
SELECT s.Product, s.Region, s.Sales_Amount, p.Category, p.Brand
FROM SalesData s
LEFT JOIN ProductDetail p ON s.Product = p.Product
''')

run_query(query)
```

```python
#Right JOIN

query = '''
SELECT s.Product, s.Region, s.Sales_Amount, p.Category, p.Brand
FROM ProductDetail p
LEFT JOIN SalesData s ON s.Product = p.Product;
'''

run_query(query)
```

```python
query = '''
SELECT s.Product, s.Region, s.Sales_Amount, p.Category, p.Brand
FROM SalesData s
LEFT JOIN ProductDetail p ON s.Product = p.Product
UNION
SELECT s.Product, s.Region, s.Sales_Amount, p.Category, p.Brand
FROM ProductDetail p
LEFT JOIN SalesData s ON s.Product = p.Product;
'''

run_query(query)
```

```python
#WHERE Clause
query = '''
SELECT Product, Region, Sales_Amount
FROM SalesData
WHERE Year = 2024 AND Sales_Amount > 1000;
'''
run_query(query)
```

```python
#GROUPBY
query = '''
SELECT p.Category, SUM(s.Sales_Amount) AS Total_Sales
FROM SalesData s
JOIN ProductDetail p ON s.Product = p.Product
GROUP BY p.Category;
'''

run_query(query)
```

```python
query = '''
SELECT p.Category, SUM(s.Sales_Amount) AS Total_Sales
FROM SalesData s
JOIN ProductDetail p ON s.Product = p.Product
WHERE s.YEAR = 2024
GROUP BY p.Category
HAVING SUM(s.Sales_Amount) > 3000;
'''


run_query(query)
```

```python
query = '''
SELECT *
FROM SalesData s
JOIN ProductDetail p ON s.Product = p.Product

'''


run_query(query)
```

```python
#ORDERBY
query = '''
SELECT s.Product, s.Year, SUM(s.Sales_Amount) as Total_Sales
FROM SalesData s
WHERE s.Year = 2024
ORDER BY Total_Sales DESC
'''

run_query(query)
```

```python
query = '''
SELECT s.Product, s.Region, SUM(s.Sales_Amount) as Total_Units_Sold
FROM SalesData s
WHERE s.Year = 2024
GROUP BY s.Product
ORDER BY Total_Units_Sold
LIMIT 1;
'''
run_query(query)
```

```python
query = '''
SELECT Region, AVG(Sales_Amount) as AVGSales, Year
FROM SalesData
GROUP BY Region
HAVING AVG(Sales_Amount) > (SELECT AVG(Sales_Amount) FROM SalesData);
'''
run_query(query)
```

```python
query = '''
SELECT Product, Sales_Amount
FROM SalesData
Where Year = 2024 AND Sales_Amount > (SELECT MAX(Sales_Amount) FROM SalesData WHERE Year = 2023)
--GROUP BY Product;
'''
run_query(query)
```

```python
query = '''
SELECT DISTINCT Product
FROM SalesData
WHERE Region IN(
  SELECT Region
  FROM SalesData
  WHERE Product = 'Laptop');
  '''
run_query(query)
```
